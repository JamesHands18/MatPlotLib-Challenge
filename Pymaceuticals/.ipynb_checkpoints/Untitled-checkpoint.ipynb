```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
df = pd.merge(mouse_metadata, study_results, on = "Mouse ID")

# Display the data table for preview
df.head()
```

```python
#number of values originally
df.shape
```

```python
#creating a new dataframe to test Mouse ID and Timepoint duplicates
test = df.drop(columns = ["Drug Regimen", "Sex", "Age_months", "Weight (g)", "Tumor Volume (mm3)","Metastatic Sites"])
test.head()
```

```python
#finds any duplicates and the index for them
test2 = pd.DataFrame(test.duplicated(), columns = ["bool"])
test2.loc[test2["bool"] == True]
```

```python
#drops the duplicate values
df.drop(index = [909, 911, 913, 915, 917], inplace = True)
```

```python
#resetting the index
df.reset_index(drop = True, inplace=True)
df.head()
```

```python
#number of values after 'cleaning'
df.shape
```

```python
#list of summary statistics for each drug regimen
mean_tumor = df.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean().to_list()
median_tumor = df.groupby("Drug Regimen")["Tumor Volume (mm3)"].median().to_list()
var_tumor = df.groupby("Drug Regimen")["Tumor Volume (mm3)"].var().to_list()
stdev_tumor = df.groupby("Drug Regimen")["Tumor Volume (mm3)"].std().to_list()
sem_tumor = df.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem().to_list()
drug_regimen = df.groupby("Drug Regimen").count().index.to_list()
```

```python
summary = pd.DataFrame({
    "Mean": mean_tumor,
    "Median": median_tumor,
    "Variance": var_tumor,
    "Standard Deviation": stdev_tumor,
    "SEM": sem_tumor
}, index = drug_regimen)
summary
```

```python
#same using the aggregate method
df.groupby("Drug Regimen")["Tumor Volume (mm3)"].agg(["mean", "median", "var", "std", "sem"])
```

```python
#max timepoint for each mouse
max_df = pd.DataFrame(df.groupby("Mouse ID")["Timepoint"].max())
max_df.insert(0, "Mouse ID", max_df.index)
max_df.reset_index(drop = True, inplace=True)
max_df
```

```python
#merged dataframe
new_df = pd.merge(df, max_df, how = "inner", on = ["Mouse ID", "Timepoint"])
new_df
```

```python
#list making
treatment_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
tumor_vol = []
```

```python
#finding the final tumor volumes for the specified drugs
for treat in treatment_list:
    for x in new_df.index:
        if new_df["Drug Regimen"][x] == treat:
            tumor_vol.append(new_df["Tumor Volume (mm3)"][x])
            tumor_vol.append(treat)

tumor_vol
```

```python
#for loop that calculates iqr and outliers for each drug
temp = []
outliers = []

for treat in treatment_list:
    for x in range(len(tumor_vol)):
        if x == (len(tumor_vol) - 1):
            break
        elif tumor_vol[x + 1] == treat:
            temp.append(tumor_vol[x])
    
    lq = np.quantile(temp, 0.25)
    uq = np.quantile(temp, 0.75)
    iqr = uq - lq
    lb = lq - (1.5 * iqr)
    ub = uq + (1.5 * iqr)
    
    for val in temp:
        if val > ub:
            outliers.append(val)
        elif val < lb:
            outliers.append(val)
    
    print(f"The interquartile range for {treat} is {iqr}")
    if len(outliers) != 0: 
        for outlier in outliers:
            print(f"One of the potential outliers for {treat} is: {outlier}")
    else:
        print(f"{treat} has no outliers")
    
    temp.clear()
    outliers.clear()
```

```python
#creates list of values for each drug
cap_l = []
ram_l = []
inf_l = []
cef_l = []

for x in range(len(tumor_vol)):
    if x == (len(tumor_vol) - 1):
        break
    elif tumor_vol[x + 1] == "Capomulin":
        cap_l.append(tumor_vol[x])
    elif tumor_vol[x + 1] == "Ramicane":
        ram_l.append(tumor_vol[x])
    elif tumor_vol[x + 1] == "Infubinol":
        inf_l.append(tumor_vol[x])
    elif tumor_vol[x + 1] == "Ceftamin":
        cef_l.append(tumor_vol[x])
```

```python
#creates box plots for each drug
plt.boxplot([cap_l, ram_l, inf_l, cef_l], labels = treatment_list);
```

```python
avg_df = df.groupby("Mouse ID").mean()
avg_df
```

```python
avg_df.plot("Weight (g)", "Tumor Volume (mm3)", kind = "scatter");
```

```python
(slope, intercept, r, p, stderr) = st.linregress(avg_df["Weight (g)"], avg_df["Tumor Volume (mm3)"])
regress_values = avg_df["Weight (g)"] * slope + intercept
line_eq = f"y = {round(slope, 2)}x + {round(intercept, 2)}"

plt.scatter(avg_df["Weight (g)"], avg_df["Tumor Volume (mm3)"]);
plt.plot(avg_df["Weight (g)"], regress_values, "r-");
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")

print(f"The linear regression model for these values is: {line_eq}")
print(f"The Pearson's r value for these values is {r}")
```
